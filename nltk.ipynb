# NLTK

O NLTK (Natural Language Toolkit) é a principal biblioteca Python de código aberto para processamento de linguagem natural (PLN), ideal para tarefas como tokenização, stemming, lematização e análise de sentimentos. Amplamente utilizada em pesquisa e academia, facilita o manuseio de textos, oferecendo acesso a corpora (bases de texto) e algoritmos essenciais.

## Introdução e preparação

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer, LancasterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag, pos_tag_sents
import string
nltk.download("stopwords")
nltk.download("punkt_tab")
nltk.download("tagsets_json")
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger")
nltk.download("maxent_ne_chunker_tab")
nltk.download("words")
```

```python
Texto = "Nós somos feitos de poeira de estrelas. Nós somos uma maneira do cosmos se autoconhecer. A imaginação nos leva a mundos que sequer existiram. Mas sem ela não vamos a lugar algum."
print(Texto)
```

## Produção de tokens

```python
sentencas = sent_tokenize(Texto, language="portuguese")
print(type(sentencas))
print(sentencas)
```

```python
print(len(sentencas))
```

```python
tokens = word_tokenize(Texto, language="portuguese")
print(tokens)
print(len(tokens))
```

## Gerenciando stop words e pontuação

```python
stops = stopwords.words("portuguese")
print(len(stops))
print(stops)
```

```python
palavras_sem_stopwords = [p for p in tokens if p not in stops]
print("Com stopwords:", Texto)
print("Sem stopwords:", palavras_sem_stopwords)
print()
print("Tamanho com stopwords:", len(Texto))
print("Tamanho sem stopwords:", len(palavras_sem_stopwords))
```

```python
print(string.punctuation)
```

```python
palavras_sem_pontuacao = [p for p in palavras_sem_stopwords if p not in string.punctuation]
print(len(palavras_sem_pontuacao))
print(Texto)
print(palavras_sem_pontuacao)
```

## Produzindo métricas

```python
frequencia = nltk.FreqDist(palavras_sem_pontuacao)
frequencia
```

```python
mais_comuns = frequencia.most_common(5)
mais_comuns
```

## Stemming na prática
Trata-se de uma técnica que reduz palavras flexionadas ou derivadas ao seu radical (raiz). O objetivo principal é mapear variações de uma palavra para uma única forma base, o que ajuda na normalização de texto e melhora a eficiência de algoritmos de busca e análise de sentimentos.

```python
stemmer = PorterStemmer()
stem1 = [stemmer.stem(word) for word in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(stem1)
```

```python
stemmer2 = SnowballStemmer("portuguese")
stem2 = [stemmer2.stem(word) for word in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(stem2)
```

```python
stemmer3 = LancasterStemmer()
stem3 = [stemmer3.stem(word) for word in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(stem3)
```

```python
token2 = sent_tokenize(Texto)

lista_tokens = []
for token_sentenca in token2:
  lista_tokens.append(word_tokenize(token_sentenca))

print(lista_tokens)

pos_sentenca = pos_tag_sents(lista_tokens)
print(pos_sentenca)
```

## Lemmatization na prática

```python
lemmatizer = WordNetLemmatizer()
resultado = [lemmatizer.lemmatize(palavra) for palavra in palavras_sem_pontuacao]
print(palavras_sem_pontuacao)
print(resultado)
```

## Busca de entidades nomeadas

```python
texto_en = "Barack Obama foi um presidente dos EUA."
token3 = word_tokenize(texto_en)
tags = pos_tag(token3)
en = nltk.ne_chunk(tags)
print(en)
```
