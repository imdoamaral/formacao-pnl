# NLTK

O NLTK (Natural Language Toolkit) é a principal biblioteca Python de código aberto para processamento de linguagem natural (PLN), ideal para tarefas como tokenização, stemming, lematização e análise de sentimentos. Amplamente utilizada em pesquisa e academia, facilita o manuseio de textos, oferecendo acesso a corpora (bases de texto) e algoritmos essenciais.

## Introdução e preparação

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer, LancasterStemmer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag, pos_tag_sents
import string
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("tagsets")
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger")
nltk.download("maxent_ne_chunker")
nltk.download("words")
```
